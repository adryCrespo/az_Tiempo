```python

import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.figure import Figure
```

```python
df = pd.read_csv("t.csv",header=None)
```

```python
df.columns = ["ciudad", "fecha", "t_min","t_max"]
```

```python
df
```

```python
df.dtypes
```

```python
datos = [row[1:] for row in df.itertuples()]
datos
```

```python
ciudades,fecha, t_min, t_max = zip(*datos)
```

```python
t_max
```

```python
# fig = Figure()
# axis = fig.add_subplot(1, 1, 1)
#Creamos la figura
f, ax = plt.subplots()
ax.set_title('Evolucion Temperaturas a lo largo del tiempo en Madrid')
ax.set_xlabel('Fecha')
ax.set_ylabel('Temperatura (C)')
ax.plot(fecha,t_min, label = "Temperatura Minima")
ax.plot(fecha, t_max, label = "Temperatura Maxima")
ax.legend(bbox_to_anchor=(1.45, 1),loc = 'upper right')
ax.tick_params(axis='x', labelsize=10, labelrotation=45)
ax.grid(color='gray', linestyle=':', linewidth=1)
;
```
